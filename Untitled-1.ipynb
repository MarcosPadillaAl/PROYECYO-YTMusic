```python


yt = YTMusic()
playlist=['RDCLAK5uy_mEXMsobutX4itJVd6YaCodxMgUV-mMtNo','PL4fGSI1pDJn4jhQB4kb9M36dvVmJQPt4T']

artistas = []
diccSolo= {}
diccSolo.clear()
colabos=[]

infoPlayList = []
for p in playlist:
     infoPlayList.append(yt.get_playlist(p))
     
#para rellenar artistas
for r in infoPlayList:
     for a in r['tracks']:
      artistas.append(a['artists'])

#for r in searchPlay['tracks']:
#     artistas.append(r['artists'])

for a in artistas:#cargamos el diccionario con los artistas que van a cantar solos
   for b in a:
    if(len(a)==1):
        diccSolo[b['name']]={'idYTMusic':b['id'],'colaboracion':[]}

```

```python
diccColabo= {}
diccColabo.clear()

for a in artistas:#aqui cargamos solo artistas que han aparecido con colaboraciones en la playlist
  if(len(a)>1):
      diccColabo[a[0]['name']]={'idYTMusic':a[0]['id'],'colaboracion':[],'subscriptores':None,'views':None}

      #if(len(a)==2):
          #dicc[b['id']]={'name':b['name']}
```

Nos quedamos con el diccionario de colaboraciones para empezar nuestro análisis. Y empezamos a meter los atributos que nos vendran en el estudio. En este caso, recorriendo otra vez los artistas, sacaremos las colaboraciones de cada uno y la añadiremos

```python
#c[1]['name']
for c in artistas:#aqui les metemos en colaboraciones sus feats
    if(len(c)>1):
      # print(c[i+1]['name'])--> funcion que recorre artistas de las colaboraciones
      #añadimos artistas a las colabos
           for i in range(len(c)-1):
            diccColabo[c[0]['name']]['colaboracion'].append(c[i+1]['name'])
#diccColabo
```

Al proseguir creando la red , nos dimos cuenta que un atributo que seria importante como el id de cada artista; algunas veces no salia. Aparecia como vacío y nos daba fallo al no poder mostrar nada. Esto era o bien porque no tienen canal creado dentro de esta plataforma o porque a veces la aplicación falla ocasionalmente. Por eso en este código buscamos estos espacios vacío, y los rellenamos usando otra funcion de la appi que nos proporciona tambien el id del artista: "search". En el caso de que no nos encuentre nada ponemos que no tiene canal creado.

```python
#para quitar los artistas que no tiene id ytMusic(fallo de la api)

for a in diccColabo:
   if(diccColabo[a]['idYTMusic']==None):
     idsvacios=yt.search(a)
     if('browseId'not in idsvacios[0]):#VER SI TIENE  ['browseId'] , SI NO LO TIENE SE PONE NO CHANNEL, SI TIENE SE PONE LO QUE ESTA EN EL ELSE  
      diccColabo[a]['idYTMusic']='No Channel'
     else:
      diccColabo[a]['idYTMusic']=idsvacios[0]['browseId']
   
#diccColabo

```

```python
#numeramos con ids 
i=0
for keys in diccColabo:
    diccColabo[keys]['id']=i
    i=i+1
```

Una vez bien creado las bases del csv, procedemos a rellenar la información de cada artista. En este caso meteremos sus visitas y sus suscriptores. Como resultado obtenemos el diccionario al completo con el que podremos crear el csv.

```python
#sacamos artistas , del diccionario por su id, sacamos los que tienen canal,y metemos su info de vistas y subscriptores
infoartista=[]
infoSubscribers=[]
infoViews=[]
for a in diccColabo:
   infoartista.append(diccColabo[a]['idYTMusic'])
  
for r in infoartista:
   if(r!='No Channel'):
      artista=yt.get_artist(r)
      infoSubscribers.append(artista['subscribers'])
      infoViews.append(artista['views'])
i=0
for keys in diccColabo:
    if(diccColabo[keys]['idYTMusic']!='No Channel'):
      diccColabo[keys]['subscriptores']=infoSubscribers[i]
      diccColabo[keys]['views']=infoViews[i]
      i=i+1    

diccColabo
```

```python
df = pd.DataFrame.from_dict(diccColabo, orient = "index",columns=['id','idYTMusic','colaboracion','subscriptores','views'])
df.to_csv('finished.csv')
df
```

```python
#para rellenar si es explicito  para artistas con colabos       
duration = []
explicit = []
publicacion=[]
vistas=[]
videosIds=[]
artistasAristas=[]
for r in infoPlayList:
     for a in r['tracks']:
          if(len(a['artists'])>1):
            explicit.append(a['isExplicit'])

#para rellenar duracion seg   para artistas con colabos     
for r in infoPlayList:
     for a in r['tracks']:
           if(len(a['artists'])>1):
            duration.append(a['duration_seconds'])
#para los artistas del csv
for r in infoPlayList:
     for a in r['tracks']:
           if(len(a['artists'])>1):
            artistasAristas.append(a['artists'])


for r in infoPlayList:
     for a in r['tracks']:
           if(len(a['artists'])>1):
               searchMusic=yt.get_song(a['videoId'])
               publicacion.append(searchMusic['microformat']['microformatDataRenderer']['publishDate'])
               vistas.append(searchMusic['microformat']['microformatDataRenderer']['viewCount'])
```

En este momento, abrimos un csv para escribir la información de las aristas.En nuestro caso se van a unir los nodos segun sus colaboraciones.Si el nodo A ha colaborado con B y C se lanzará una arista desde A hasta B y desde A hasta C.
Tambien hemos añadido atributos complementarios para su posterior estudio en Gephi, como la duracion de la cancion en cuestión en la que colaboran, si la letra de la cancion tiene contenido explicito o no y la fecha de la publicación de la cancion en ytMusic.

```python
header = ['artista', 'colaboracion','duracion','explicito','fecha_publicacion','visitas']
writedicc = {}
with open('aristas.csv','w',encoding='utf-8',newline='') as f:
    writer= csv.writer(f)
    writer.writerow(header)
    iter=0
    repe=0
    for r in artistasAristas:
        for i in range(len(r)-1):
            writer.writerow([r[0]['name'],r[i+1]['name'],duration[iter],explicit[iter],publicacion[iter],vistas[iter]])
        iter=iter+1
            
      # iter=iter+1
          
          
```

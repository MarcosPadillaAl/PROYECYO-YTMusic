Al proseguir creando la red , nos dimos cuenta que un atributo que seria importante como el id de cada artista; algunas veces no salia. Aparecia como vacío y nos daba fallo al no poder mostrar nada. Por eso en este código buscamos estos espacios vacío, y los rellenamos usando otra funcion de la appi que nos proporciona tambien el id del artista: "search". En el caso de que no nos encuentre nada ponemos que no tiene canal creado.

```python
#para quitar los artistas que no tiene id ytMusic(fallo de la api)

for a in diccColabo:
   if(diccColabo[a]['idYTMusic']==None):
     idsvacios=yt.search(a)
     if('browseId'not in idsvacios[0]):#VER SI TIENE  ['browseId'] , SI NO LO TIENE SE PONE NO CHANNEL, SI TIENE SE PONE LO QUE ESTA EN EL ELSE  
      diccColabo[a]['idYTMusic']='No Channel'
     else:
      diccColabo[a]['idYTMusic']=idsvacios[0]['browseId']
   
#diccColabo

```

```python
#numeramos con ids 
i=0
for keys in diccColabo:
    diccColabo[keys]['id']=i
    i=i+1
```

Una vez bien creado las bases del csv, procedemos a rellenar la información de cada artista. En este caso meteremos sus visitas y sus suscriptores. Como resultado obtenemos el diccionario al completo con el que podremos crear el csv.

```python
#sacamos artistas , del diccionario por su id, sacamos los que tienen canal,y metemos su info de vistas y subscriptores
infoartista=[]
infoSubscribers=[]
infoViews=[]
for a in diccColabo:
   infoartista.append(diccColabo[a]['idYTMusic'])
  
for r in infoartista:
   if(r!='No Channel'):
      artista=yt.get_artist(r)
      infoSubscribers.append(artista['subscribers'])
      if(artista['views']!=None):
         infoViews.append(artista['views'])
      else:
         infoViews.append("No info")
i=0
for keys in diccColabo:
    if(diccColabo[keys]['idYTMusic']!='No Channel'):
      diccColabo[keys]['subscriptores']=infoSubscribers[i]
      diccColabo[keys]['views']=infoViews[i]
      i=i+1    
```

Con el diccionario obtenido arriba, tenemos colaboraciones completadas, pero no los nodos. El nodo a colabora con b y c, pero solo esta reconocido como nodo a , b y c no. Tenemos que crear nodos para los colaboradores tambien. Esta estructura nos viene bien para sacar los csv de aristas y de nodos para el análisis de gephi, porque solo tiene que tratar las colaboraciones una única vez. Es decir que va a tratar como 1 colaboración A-B. De la otra manera sería muy costoso porque trataría las colaboraciones 2 veces A-B y B-A

```python
#para rellenar si es explicito  para artistas con colabos       
duration = []
explicit = []
publicacion=[]
vistas=[]
videosIds=[]
artistasAristas=[]
for r in infoPlayList:
     for a in r['tracks']:
          if(len(a['artists'])>1):
            explicit.append(a['isExplicit'])

#para rellenar duracion seg   para artistas con colabos     
for r in infoPlayList:
     for a in r['tracks']:
           if(len(a['artists'])>1):
            duration.append(a['duration_seconds'])
#para los artistas del csv
for r in infoPlayList:
     for a in r['tracks']:
           if(len(a['artists'])>1):
            artistasAristas.append(a['artists'])

#para las vistas de cada cancion
for r in infoPlayList:
     for a in r['tracks']:
           if(len(a['artists'])>1):
               searchMusic=yt.get_song(a['videoId'])
               publicacion.append(searchMusic['microformat']['microformatDataRenderer']['publishDate'])
               vistas.append(searchMusic['microformat']['microformatDataRenderer']['viewCount'])
```

Utilizamos la tabla de aristas para sacar los artistas que no han sido reconocidos como nodos y lo añadimos vacios

```python

for r in artistasAristas:
   for i in range(len(r)-1):
        a=i+1
        if(r[a]['name'] not in diccColabo):
           diccColabo[r[a]['name']]={'idYTMusic':'None','colaboracion':[],'subscriptores':None,'views':None,'id':None}
           search_results=yt.search(r[a]['name'])
           
           if('browseId'in search_results[0]):
            if(search_results[0]['browseId'][0]=='U'):
               
                searcArtist=yt.get_artist(search_results[0]['browseId'])
                diccColabo[r[a]['name']]['idYTMusic']=search_results[0]['browseId']
                diccColabo[r[a]['name']]['subscriptores']=searcArtist['subscribers']
                if(searcArtist['views']==None):
                     diccColabo[r[a]['name']]['views']="No info"
                else:
                    diccColabo[r[a]['name']]['views']=searcArtist['views']
                for c in  diccColabo:
                    for i in range(len(diccColabo[c]['colaboracion'])):
                        if(r[a]['name'] in diccColabo[c]['colaboracion'][i]):
                            diccColabo[r[a]['name']]['colaboracion'].append(c)
i=0
for keys in diccColabo:
    if(diccColabo[keys]['id']==None):
        diccColabo[keys]['id']=i
    i=i+1           
```
